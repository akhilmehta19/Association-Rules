```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_csv("C:/Users/Akhil/Desktop/ASSIGNMENTS/my_movies.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df1 = pd.get_dummies(df)
df1
```

```python
frequent = apriori(df1,min_support=0.2,use_colnames=True)
frequent
```

```python
rules = association_rules(frequent,metric="lift",min_threshold=0.8)
rules
```

```python
rules.sort_values("lift",ascending=False)[0:100]
```

```python
rules[rules.lift>3]
```
